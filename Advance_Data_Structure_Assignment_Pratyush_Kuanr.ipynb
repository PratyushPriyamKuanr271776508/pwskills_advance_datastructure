```python
### 1. Code to reverse a string:

def reverse_string(s):
    return s[::-1]

# Example usage:
print(reverse_string("hello"))  # Output: "olleh"



### 2. Code to count the number of vowels in a string:

def count_vowels(s):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char in vowels)

# Example usage:
print(count_vowels("hello"))  # Output: 2



### 3. Code to check if a given string is a palindrome:

def is_palindrome(s):
    return s == s[::-1]

# Example usage:
print(is_palindrome("madam"))  # Output: True




### 4. Code to check if two given strings are anagrams:

def are_anagrams(s1, s2):
    return sorted(s1) == sorted(s2)

# Example usage:
print(are_anagrams("listen", "silent"))  # Output: True




### 5. Code to find all occurrences of a given substring within another string:

def find_substring_occurrences(main_str, sub_str):
    return [i for i in range(len(main_str) - len(sub_str) + 1) if main_str[i:i+len(sub_str)] == sub_str]

# Example usage:
print(find_substring_occurrences("banana", "ana"))  # Output: [1, 3]


### 6. Code for basic string compression:


def compress_string(s):
    compressed = []
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1
    compressed.append(s[-1] + str(count))  # Last character
    return ''.join(compressed)

# Example usage:
print(compress_string("aaabbccc"))  # Output: "a3b2c3"



### 7. Code to check if a string has all unique characters:


def has_unique_chars(s):
    return len(set(s)) == len(s)

# Example usage:
print(has_unique_chars("abcde"))  # Output: True



### 8. Code to convert a string to uppercase or lowercase:


def convert_case(s, to_upper=True):
    return s.upper() if to_upper else s.lower()

# Example usage:
print(convert_case("Hello", to_upper=True))  # Output: "HELLO"
print(convert_case("Hello", to_upper=False))  # Output: "hello"



### 9. Code to count the number of words in a string:


def count_words(s):
    return len(s.split())

# Example usage:
print(count_words("This is a test string"))  # Output: 5



### 10. Code to concatenate two strings without using the `+` operator:


def concatenate_strings(s1, s2):
    return ''.join([s1, s2])

# Example usage:
print(concatenate_strings("hello", "world"))  # Output: "helloworld"



### 11. Code to remove all occurrences of a specific element from a list:


def remove_all_occurrences(lst, element):
    return [x for x in lst if x != element]

# Example usage:
print(remove_all_occurrences([1, 2, 3, 2, 4], 2))  # Output: [1, 3, 4]



### 12. Code to find the second largest number in a list:


def second_largest(lst):
    unique_lst = list(set(lst))  # Remove duplicates
    unique_lst.sort()
    return unique_lst[-2] if len(unique_lst) >= 2 else None

# Example usage:
print(second_largest([1, 2, 3, 4, 5]))  # Output: 4



### 13. Code to count occurrences of each element in a list:


def count_occurrences(lst):
    return {x: lst.count(x) for x in set(lst)}

# Example usage:
print(count_occurrences([1, 2, 2, 3, 3, 3]))  # Output: {1: 1, 2: 2, 3: 3}



### 14. Code to reverse a list in-place:


def reverse_list(lst):
    left, right = 0, len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1

# Example usage:
lst = [1, 2, 3, 4]
reverse_list(lst)
print(lst)  # Output: [4, 3, 2, 1]



### 15. Code to remove duplicates from a list while preserving order:


def remove_duplicates(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]

# Example usage:
print(remove_duplicates([1, 2, 2, 3, 3, 1]))  # Output: [1, 2, 3]



### 16. Code to check if a list is sorted:


def is_sorted(lst):
    return lst == sorted(lst) or lst == sorted(lst, reverse=True)

# Example usage:
print(is_sorted([1, 2, 3]))  # Output: True
print(is_sorted([3, 2, 1]))  # Output: True
print(is_sorted([1, 3, 2]))  # Output: False



### 17. Code to merge two sorted lists:


def merge_sorted_lists(lst1, lst2):
    return sorted(lst1 + lst2)

# Example usage:
print(merge_sorted_lists([1, 3, 5], [2, 4, 6]))  # Output: [1, 2, 3, 4, 5, 6]



### 18. Code to find the intersection of two lists:


def list_intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))

# Example usage:
print(list_intersection([1, 2, 3], [2, 3, 4]))  # Output: [2, 3]



### 19. Code to find the union of two lists without duplicates:


def list_union(lst1, lst2):
    return list(set(lst1) | set(lst2))

# Example usage:
print(list_union([1, 2, 3], [2, 3, 4]))  # Output: [1, 2, 3, 4]



### 20. Code to shuffle a list without using built-in functions:


import random

def shuffle_list(lst):
    for i in range(len(lst) - 1, 0, -1):
        j = random.randint(0, i)
        lst[i], lst[j] = lst[j], lst[i]

# Example usage:
lst = [1, 2, 3, 4, 5]
shuffle_list(lst)
print(lst)  # Output: List shuffled randomly



### 21. Code to find common elements between two tuples:


def tuple_intersection(t1, t2):
    return tuple(set(t1) & set(t2))

# Example usage:
print(tuple_intersection((1, 2, 3), (2, 3, 4)))  # Output: (2, 3)



### 22. Code to find the intersection of two sets entered by the user:


def find_set_intersection():
    set1 = set(map(int, input("Enter first set: ").split(',')))
    set2 = set(map(int, input("Enter second set: ").split(',')))
    print("Intersection:", set1 & set2)

# Example usage:
# Input: "1,2,3" and "2,3,4"
# Output: "Intersection: {2, 3}"

find_set_intersection()

### 23. Code to concatenate two tuples:


def concatenate_tuples(t1, t2):
    return t1 + t2

# Example usage:
print(concatenate_tuples((1, 2), (3, 4)))  # Output: (1, 2, 3, 4)

### 24. Code to find elements in the first set but not in the second:


def difference_between_sets():
    set1 = set(input("Enter first set of strings (comma-separated): ").split(','))
    set2 = set(input("Enter second set of strings (comma-separated): ").split(','))
    difference = set1 - set2
    print("Elements in the first set but not in the second:", difference)

# Example usage:
# Input: "apple,banana,grape" and "banana,orange"
# Output: {"apple", "grape"}



### 25. Code to return elements within a range of indices from a tuple:


def tuple_range_elements(t, start, end):
    return t[start:end+1]

# Example usage:
t = (10, 20, 30, 40, 50)
start, end = 1, 3
print(tuple_range_elements(t, start, end))  # Output: (20, 30, 40)



### 26. Code to find the union of two sets of characters:


def union_of_sets():
    set1 = set(input("Enter first set of characters (comma-separated): ").split(','))
    set2 = set(input("Enter second set of characters (comma-separated): ").split(','))
    union_set = set1 | set2
    print("Union of both sets:", union_set)

# Example usage:
# Input: "a,b,c" and "b,c,d"
# Output: {"a", "b", "c", "d"}



### 27. Code to return the maximum and minimum values from a tuple using tuple unpacking:


def max_min_from_tuple(t):
    return max(t), min(t)

# Example usage:
t = (3, 1, 4, 1, 5, 9)
print(max_min_from_tuple(t))  # Output: (9, 1)



### 28. Code to find the union, intersection, and difference of two sets:


def set_operations(set1, set2):
    union_set = set1 | set2
    intersection_set = set1 & set2
    difference_set = set1 - set2
    print(f"Union: {union_set}, Intersection: {intersection_set}, Difference: {difference_set}")

# Example usage:
set1 = {1, 2, 3}
set2 = {2, 3, 4}
set_operations(set1, set2)
# Output: Union: {1, 2, 3, 4}, Intersection: {2, 3}, Difference: {1}



### 29. Code to count occurrences of an element in a tuple:


def count_occurrences_in_tuple(t, element):
    return t.count(element)

# Example usage:
t = (1, 2, 2, 3, 4, 2)
element = 2
print(count_occurrences_in_tuple(t, element))  # Output: 3



### 30. Code to find the symmetric difference of two sets of strings:


def symmetric_difference_between_sets():
    set1 = set(input("Enter first set of strings (comma-separated): ").split(','))
    set2 = set(input("Enter second set of strings (comma-separated): ").split(','))
    sym_diff = set1 ^ set2
    print("Symmetric difference of both sets:", sym_diff)

# Example usage:
# Input: "apple,banana" and "banana,grape"
# Output: {"apple", "grape"}

symmetric_difference_between_sets()

### 31. Code to return a dictionary with word frequencies from a list of words:


def word_frequencies(words):
    freq_dict = {}
    for word in words:
        if word in freq_dict:
            freq_dict[word] += 1
        else:
            freq_dict[word] = 1
    return freq_dict

# Example usage:
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
print(word_frequencies(words))
# Output: {'apple': 3, 'banana': 2, 'orange': 1}



### 32. Code to merge two dictionaries, adding values for common keys:


def merge_dictionaries(dict1, dict2):
    merged = dict1.copy()
    for key, value in dict2.items():
        if key in merged:
            merged[key] += value
        else:
            merged[key] = value
    return merged

# Example usage:
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
print(merge_dictionaries(dict1, dict2))
# Output: {'a': 10, 'b': 25, 'c': 15}



### 33. Code to access a value in a nested dictionary using a list of keys:


def get_nested_value(d, keys):
    for key in keys:
        if isinstance(d, dict) and key in d:
            d = d[key]
        else:
            return None
    return d

# Example usage:
nested_dict = {'a': {'b': {'c': 42}}}
keys = ['a', 'b', 'c']
print(get_nested_value(nested_dict, keys))  # Output: 42



### 34. Code to return a sorted dictionary based on values:


def sort_dict_by_values(d, reverse=False):
    return dict(sorted(d.items(), key=lambda item: item[1], reverse=reverse))

# Example usage:
d = {'a': 3, 'b': 1, 'c': 2}
print(sort_dict_by_values(d))  # Output: {'b': 1, 'c': 2, 'a': 3}
print(sort_dict_by_values(d, reverse=True))  # Output: {'a': 3, 'c': 2, 'b': 1}



### 35. Code to invert a dictionary, handling multiple keys with the same value:


def invert_dictionary(d):
    inverted = {}
    for key, value in d.items():
        if value in inverted:
            inverted[value].append(key)
        else:
            inverted[value] = [key]
    return inverted

# Example usage:
d = {'a': 1, 'b': 2, 'c': 1}
print(invert_dictionary(d))
# Output: {1: ['a', 'c'], 2: ['b']}
```
